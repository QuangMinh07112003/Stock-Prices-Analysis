```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=vnindex_df_plotly['Date'], y=vnindex_df_plotly['Close'],
                    mode='lines',
                    name='lines'))
fig.add_trace(go.Scatter(x=vnindex_df_plotly['Date'], y=vnindex_df_plotly['rolling_mean50'],
                    mode='lines', #+markers
                    name='lines'))
fig.add_trace(go.Scatter(x=vnindex_df_plotly['Date'], y=vnindex_df_plotly['rolling_mean200'],
                    mode='lines',
                    name='lines'))

fig.show()
```

```python
vnindex_df_Q_plotly = vnindex_df.resample('Q').last().reset_index()
fig = px.line(vnindex_df_Q_plotly, x="Date", y="Close", title='VNINDEX price')
fig.show()

```

```python
vnindex_df_Q_bar = vnindex_df_Q_plotly.set_index('Date').loc['2015-01-01':].pct_change().reset_index()
vnindex_df_Q_bar.head(5)
```

```python
import plotly.express as px

fig = px.bar(vnindex_df_Q_bar, x='Date', y='Close', text=[f'{i:.1%}' for i in vnindex_df_Q_bar['Close']]) #, orientation='h'
fig.update_traces(textfont_size=16, textangle=0, )#, marker_color=colors
fig.show()
```

```python
# Create custom colorscale

# colorname:colorscale
color_mapping={
    'red': 'rgb(255,59,48)',
    'green': 'rgb(52,199,89)',
    'indigo': 'rgb(88,86,214)',
}
```

```python
vnindex_df_Q_bar['color'] = vnindex_df_Q_bar['Close'].apply(lambda x: 'red' if x<0 else 'green')
vnindex_df_Q_bar['color_code'] = vnindex_df_Q_bar['color'].map(color_mapping)
vnindex_df_Q_bar.tail()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Bar(
    x=vnindex_df_Q_bar['Date'],
    y=vnindex_df_Q_bar['Close'],
    text=[f'{i:.1%}' for i in vnindex_df_Q_bar['Close']],
    marker_color=vnindex_df_Q_bar['color_code'], # marker color can be a single color value or an iterable

)])
fig.update_traces(textfont_size=16, textangle=0, )
fig.update_layout(title_text='Return quarterly', template="ggplot2") # "simple_white" plotly_white
```

**Monthly**

```python
vnindex_df_M_plotly = vnindex_df.resample('M').last().reset_index()
vnindex_df_M_bar = vnindex_df_M_plotly.set_index('Date').loc['2016-01-01':].pct_change()
vnindex_df_M_bar2016 = vnindex_df_M_bar.loc['2016'].reset_index()
vnindex_df_M_bar2017 = vnindex_df_M_bar.loc['2017'].reset_index()
vnindex_df_M_bar2018 = vnindex_df_M_bar.loc['2018'].reset_index()

```

```python

from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=3, cols=1)

fig.append_trace(go.Scatter(
    x=vnindex_df_M_bar2016['Date'],
    y=vnindex_df_M_bar2016['Close'],
), row=1, col=1)

fig.append_trace(go.Scatter(
    x=vnindex_df_M_bar2017['Date'],
    y=vnindex_df_M_bar2017['Close'],
), row=2, col=1)

fig.append_trace(go.Scatter(
    x=vnindex_df_M_bar2018['Date'],
    y=vnindex_df_M_bar2018['Close'],
), row=3, col=1)


fig.update_layout(height=600, width=800, title_text="Stacked Subplots")
fig.show()
```
